```python
import pandas as pd

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
print(os.listdir("../input/movielens-20m-dataset"))
```

```python
# import movie data set and look at columns
movie = pd.read_csv("../input/movielens-20m-dataset/movie.csv")
movie.columns
```

```python
# what we need are movie id and title columns
movie = movie.loc[:,["movieId","title"]]
movie.head(10)
```

```python
# import rating data and look at columns
rating = pd.read_csv("../input/movielens-20m-dataset/rating.csv")
rating.columns
```

```python
# what we need are user id, movie id and rating columns
rating = rating.loc[:,["userId","movieId","rating"]]
rating.head(10)
```

```python
# then merge movie and rating data
data = pd.merge(movie,rating)
```

```python
# now lets look at our data 
data.head(10)
```

```python
data.shape
```

```python
data = data.iloc[:1000000,:]
```

```python
data.shape
data.head()
```

```python
# Lets make a pivot table in order to make rows are users and columns are movies, the value in this table are rating given by the user to this certain movie
pivot_table = data.pivot_table(index = ["userId"],columns = ["title"],values = "rating")
pivot_table.head(10)
```

```python
movie_watched = pivot_table["Bad Boys (1995)"]
movie_watched
```

```python
movie_watched = pivot_table["Bad Boys (1995)"]
# find correlation between "Bad Boys (1995)" and other movies
similarity_with_other_movies = pivot_table.corrwith(movie_watched)
# sort the similarities from the highest to the lowest
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
# output the top 5 results
similarity_with_other_movies.head(6)
```

```python
movie_watched = pivot_table["Two Bits (1995)"]
# find correlation between "Two Bits (1995)" and other movies
similarity_with_other_movies = pivot_table.corrwith(movie_watched)
# sort the similarities from the highest to the lowest
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)
# output the top 6 results
similarity_with_other_movies.head(6)
```
